## Canny edge detection

> The Canny edge detector is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images.

See: [Canny edge detector](https://en.wikipedia.org/wiki/Canny_edge_detector) on Wikipedia.


OpenCV Tutorial: http://docs.opencv.org/master/doc/py_tutorials/py_imgproc/py_canny/py_canny.html#canny

## Template matching

> Template Matching is a method for searching and finding the location of a template image in a larger image

OpenCV Tutorial: http://docs.opencv.org/master/doc/py_tutorials/py_imgproc/py_template_matching/py_template_matching.html#py-template-matching


```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import rcParams as pltSettings
import itertools
from IPython.display import Image

pltSettings['figure.figsize'] = (12.0, 24.0)
pltSettings['figure.dpi'] = 160
```

```python
FILE_BASENAME = 'Everytown-for-Gun-Safety-AF-1'
FILE_OUT_EXT = 'tiff'

orig_filename = '{}.ppm'.format(FILE_BASENAME)
orig_img = cv2.imread(orig_filename,0)
```

### Thresholding

```python
ret,thresh_binary = cv2.threshold(orig_img,127,255,cv2.THRESH_BINARY)
thresh_mean = cv2.adaptiveThreshold(orig_img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
thresh_gaussian = cv2.adaptiveThreshold(orig_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)

# Otsu's thresholding after Gaussian filtering
blur = cv2.GaussianBlur(orig_img,(5,5),0)
ret3,otsu = cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

thresholds = {
    'binary': {
        'thresh': thresh_binary,
        'title': 'Binary Thresh'
    },
    'mean': {
        'thresh': thresh_mean,
        'title': 'Mean Thresh'
    },
    'gaussian': {
        'thresh': thresh_gaussian,
        'title': 'Gaussian Thresh'
    },
    'otsu': {
        'thresh': otsu,
        'title': 'Otsu-Gaussian Thresh'
    }
}
```

### Find edges, lines

```python
for key, item in thresholds.iteritems ():
    edges = cv2.Canny(item.get('thresh'),50,150)
    item['edges'] = edges
    lines = cv2.HoughLinesP(item['edges'], 1, np.pi/180.0, 2, np.array([]), 80, 1)
#     lines = cv2.HoughLines(item['edges'],1,np.pi/180,200)
        
    item['lines'] = lines
```

### Draw found lines onto an image

```python
RED = (0,0,255)
BLUE = (255,0,0)
GREEN = (0,255,0)
color_cycle = itertools.cycle([RED, BLUE, GREEN])

for key, item in thresholds.iteritems():
    line_img = cv2.cvtColor(item['thresh'], cv2.cv.CV_GRAY2BGR)

    for line in item['lines'].squeeze():
        start_pt = tuple(int(x) for x in line[:2])
        end_pt = tuple(int(x) for x in line[2:])
        cv2.line(line_img, start_pt, end_pt, RED, 2)
    
    item['image'] = line_img
    item['outfile'] = '{}-{}.{}'.format(FILE_BASENAME, key, FILE_OUT_EXT)
```

```python
rect_list = []
otsu = thresholds.get('otsu')
otsu_img = otsu['thresh']
lines_list = otsu['lines'][0]
rectangles_img = cv2.cvtColor(otsu_img, cv2.cv.CV_GRAY2BGR)

for line in lines_list:
    x1,y1,x2,y2 = line
    vertex1 = (x1, y1)
    vertex2 = (x2, (y1-20))
    rect_list.append(cv2.rectangle(rectangles_img, vertex1, vertex2, GREEN, 2))

    
thresholds['otsu-rect'] = {
        'thresh': otsu,
        'title': 'Otsu Thresh Rectangles',
        'image': rectangles_img,
        'rectangles': rect_list,
        'outfile': '{}-{}.{}'.format(FILE_BASENAME, 'otsu-rect', FILE_OUT_EXT)
}

print(len(lines_list))
for n,line in enumerate(lines_list):
    x1,y1,x2,y2 = line
    vertex1 = (x1, y1)
    vertex2 = (x2, (y1-20))
    cropped = orig_img[(y1-24):(y1+5), (x1-50):(x2+50)]
#     print((y1-20), y1, x1, x2)
#     print(n, cropped)
    img_name = '{}-{}-box{}.{}'.format(FILE_BASENAME, 'orig-rect', n, 'ppm')
    cv2.imwrite(img_name, cropped)
```

### Set up images for display, export

```python
images = [item for item in thresholds.itervalues()]
```

#### Export images

```python
for n, img in enumerate(images):
    print("Exporting: {}".format(images[n]['outfile']))
    cv2.imwrite(images[n]['outfile'], images[n]['image'])
```

#### Display images

```python
images_len = len(images)

for n, img in enumerate(images):
    plt.subplot(3,2,n+1),plt.imshow(images[n]['image'],'gray')
    plt.title(images[n]['title'])
    plt.xticks([]),plt.yticks([])
plt.show()
```
